```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/cumulative.csv")

df['koi_disposition'].value_counts()

df['koi_prad'].hist(bins = 50, log = True)
plt.figure()

df['koi_period'].hist(bins = 100, log = True)
plt.figure()

df['koi_teq'].hist(bins = 50, log = True)
plt.figure()

df.plot.scatter(x = 'koi_period', y = 'koi_prad', alpha = 0.5)
df.plot.scatter(x = 'koi_srad', y = 'koi_prad', alpha = 0.5)

df.isna().sum().sort_values(ascending = False).head(15)

df.describe()
```
