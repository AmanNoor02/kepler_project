```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix
df = pd.read_csv("../data/cumulative.csv")

y = df["koi_disposition"]
x = df[["koi_period", "koi_prad", "koi_depth", "koi_srad", "koi_steff", "koi_slogg"]]

x = x.dropna()
y = y.loc[x.index]

x.isnull().sum()
(x.isnull().sum() / len(x)) * 100

le = LabelEncoder()

y_encoded = le.fit_transform(y)

print(dict(zip(le.classes_, le.transform(le.classes_))))

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled = pd.DataFrame(x_scaled, columns = x.columns, index = x.index)
print(x_scaled)

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_encoded, test_size = 0.2, random_state = 73, stratify = y_encoded)

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

log_reg_scaled = LogisticRegression(max_iter = 1000, random_state = 73, class_weight= 'balanced')
log_reg_scaled.fit(x_train_scaled, y_train)
y_pred_scaled = log_reg_scaled.predict(x_test_scaled)

accuracy_scaled = accuracy_score(y_test, y_pred_scaled)
print("Scaled Model Accuracy:", accuracy_scaled)
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_scaled, target_names=['CANDIDATE', 'CONFIRMED', 'FALSE POSITIVE']))

cm_scaled = confusion_matrix(y_test, y_pred_scaled)
sns.heatmap(
    cm_scaled, annot = True, fmt = 'd', cmap = 'Blues', xticklabels=['CANDIDATE', 'CONFIRMED', 'FALSE POSITIVE'], yticklabels=['CANDIDATE', 'CONFIRMED', 'FALSE POSITIVE'])
plt.title("Confusion Matrix (Scaled Data)")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```
